```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv("/content/Customers.csv")
df.sample(3)
```

```python
df
```

```python
#Deleted empty and repeated value columns
x = ['Prefix', 'Unnamed: 5','Unnamed: 6', 'Unnamed: 17', 'Unnamed: 18', 'BirthDate.1', 'Unnamed: 8']
df.drop(x, axis = 1, inplace = True)
df.head(3)
```

```python
#Renaming Unnamed Column to Full Name
df.rename(columns={"Unnamed: 4" : "Full Name"}, inplace = True)
df.head(3)
```

```python
df.MaritalStatus = df.MaritalStatus.replace("M", "Married")
df.MaritalStatus = df.MaritalStatus.replace("S", "Single")
df.Gender = df.Gender.replace("M", "Male")
df.Gender = df.Gender.replace("F", "Female")
df.HomeOwner = df.HomeOwner.replace("Y", "Yes")
df.HomeOwner = df.HomeOwner.replace("N", "No")
df.sample(3)
```

```python
df.isnull().sum()
```

```python
# Using the mode which is the "Male" to fill up the NA values
df.fillna(df.dtypes.replace({'object' : 'Male'}), inplace = True)
df.isnull().sum()
```

**Count and Bar chart for Marital status**

```python
df.MaritalStatus.value_counts().reset_index()
```

```python
df.MaritalStatus.value_counts().plot(kind = 'bar', color = ["Blue","Red"])
plt.title("Marital Status")
plt.xlabel("Status")
plt.ylabel("Counts")
```

**Visualizing customer's gender using pie chart**

```python
df.Gender.value_counts().plot(kind = 'pie')
plt.title("Gender")
```
